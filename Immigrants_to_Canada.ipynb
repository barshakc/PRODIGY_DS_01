```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
path="C:\\Users\\HP\\Downloads\\canadian_immegration_data.csv"
df=pd.read_csv(path)
```

```python
#making 'Country' column as index column
df.set_index('Country',inplace=True)
```

```python
df.head()
```

```python
years=list(map(str,range(1980,2014)))
print(years)
```

```python
#Plotting Bar Chart for the top 5 countries migrating to Canada from year(1980-2013)
df.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df1=df.head()
df1=df1[years].transpose()
df1.head()
```

```python
df1.plot(kind='area',figsize=(20,10))
plt.title('Immigration of Top 5 countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
#Plotting the total immigrants from country 'Afghanistan' to Canada in a box plot
df2 = df.loc[['Afghanistan'], years].transpose()
df2.head()
```

```python
df2.plot(kind='box', figsize=(10,8))
plt.title('Afghanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')
plt.show()
```
