### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
path="C:\\Users\\HP\\Downloads\\canadian_immegration_data.csv"
df=pd.read_csv(path)
```

```python
df.describe()
```

```python
df.info()
```

```python
#setting 'Country' column as the index column
df.set_index('Country',inplace=True)
```

```python
df.head()
```

```python
years=list(map(str,range(1980,2014)))
print(years)
```

```python
#Creating a dataframe series for the top 5 countries migrating to Canada from year(1980-2013)
df.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df1=df.head()
df1=df1[years].transpose()
df1.head()
```

#### Area Plot

```python
df1.plot(kind='area',figsize=(20,10))
plt.title('Immigration of Top 5 countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

##### Findings : India has the largest number of people migrating to Canada over the years

```python
#Plotting the total immigrants from country 'India' to Canada in a bar chart
df2 = df.loc[['India'], years].transpose()
df2.head()
```

```python
df2.plot(kind='bar', figsize=(14,8),color='green')
plt.title('Indian Immigrants from 1980 - 2013')
plt.xlabel('India')
plt.ylabel('Number of Immigrants')
plt.show()
```

#### Findings : India had the most number of immigrants in the year '2005'

```python
# Creating a dataframe series for top 10 countries in descending order of Immigrants
df.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df1=df.head(10)
df1=df1['Total'].transpose()
df1.head(10)
```

#### Bar Chart

```python
#Plotting a bar chart to show the total immigrants from top 10 countries to Canada
df1.plot(kind='bar',figsize=(8,6))
plt.title('Total Immigrants from Top 10 countries to Canada')
plt.ylabel('Total Number of Immigrants')
plt.xlabel('Countries')
plt.show()
```

```python
df2=df.groupby('Continent',axis=0).sum()
```

```python
df2['Total'].plot(kind='bar',figsize=(8,6))
plt.title('Total Immigrants From Different Continents')
plt.xlabel('Continent')
plt.ylabel('Total')
plt.show()
```

##### Findings: Asia has the largest immigrants to Canada
